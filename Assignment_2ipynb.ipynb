```python
6.#List all categorical and numerical features separately.
categorical_features = df.select_dtypes(include=['object']).columns
numerical_features = df.select_dtypes(include=['int64', 'float64']).columns
print("Categorical Features:")
print(categorical_features)
print("\nNumerical Features:")
print(numerical_features)
print("\nTotal Categorical Features:", len(categorical_features))
print("Total Numerical Features:", len(numerical_features))
```

```python
7.#Plot the distribution of customer ages using a histogram.
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'], bins=20, kde=True ,color='pink',edgecolor='black')
plt.title('Distribution of Customer Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
8.#Create a count plot showing the number of customers in each occupation.
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Occupation', palette='viridis')
plt.title('Number of Customers in Each Occupation')
plt.xlabel('Occupation')
plt.ylabel('Count')
plt.show()
```

```python
9#.Plot the distribution of monthly income categories.
plt.figure(figsize=(10, 6))
sns.countplot(x='Monthly Income',hue='Monthly Income',data=df,palette="Set3",legend=False)
plt.title('Distribution of Monthly Income Categories')
plt.xlabel('Monthly Income Category')
plt.ylabel('Count')
plt.show()
```

```python
10.#Find the top 5 most common educational qualifications.
df['Educational Qualifications'].value_counts().head(5)
```

```python
11.#Compare the distribution of Gender across different Occupations using an appropriate visualization.
plt.figure(figsize=(10, 6))
sns.countplot(x='Gender', hue='Occupation', data=df, palette='viridis')
plt.title('Distribution of Gender Across Different Occupations')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
12.#Analyze the relationship between Marital Status and Feedback.
plt.figure(figsize=(10, 6))
sns.countplot(x='Marital Status', hue='Feedback', data=df, palette='dark:brown')
plt.xlabel('Marital Status')
plt.ylabel('Count')
plt.title('Relationship between Marital Status and Feedback')
plt.show()
```

```python
13.#Compare Family Size across different Monthly Income groups using a box plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df,x='Monthly Income',hue='Monthly Income',y='Family size',palette='pastel',legend=False)
plt.xlabel('Family Size')
plt.ylabel('Monthly Income')
plt.title('Comparison of Family Size Across Different Monthly Income Groups')
plt.show()
```

```python
14.#Visualize the relationship between Age and Family Size using a scatter plot.
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Age', y='Family size', hue='Age', palette='viridis')
plt.xlabel('Age')
plt.ylabel('Family Size')
plt.title('Relationship between Age and Family Size')
plt.show()
```

```python
15.#Create a heatmap of correlations among all numerical features.
numerical_features = df.select_dtypes(include=['int64', 'float64']).columns
correlation_matrix = df[numerical_features].corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
16.#Analyze how Occupation, Monthly Income, and Feedback are related.
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Occupation', y='Monthly Income', hue='Feedback', palette='viridis')
plt.xlabel('Occupation')
plt.ylabel('Monthly Income')
plt.title('Relationship between Occupation, Monthly Income, and Feedback')
plt.show()
```

```python
17.#Compare the average family size for each educational qualification.
sns.barplot(data=df, x='Educational Qualifications', y='Family size', palette='viridis')
plt.xlabel('Educational Qualification')
plt.ylabel('Average Family Size')
plt.title('Average Family Size for Each Educational Qualification')
plt.show()
```

```python
18.#Plot customer locations using Latitude and Longitude on a scatter plot and identify any visible clusters.
sns.scatterplot(data=df,x='latitude',y='longitude',color='blue')
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.title('Customer Locations')
plt.show()
```

```python
#Perform a complete EDA and summarize at least five meaningful business insights
# **************************
#  Complete EDA Summary
# *************************

print("Dataset Shape:")
print(df.shape)

print("\nDataset Info:")
print(df.info())

print("\nMissing Values:")
print(df.isnull().sum())

print("\nSummary Statistics:")
print(df.describe())

print("\nCategorical Features Summary:")
print(df.describe(include='object'))

# Top Occupations
print("\nTop Occupations:")
print(df["Occupation"].value_counts())

# Monthly Income Distribution
print("\nMonthly Income Distribution:")
print(df["Monthly Income"].value_counts())

# Feedback Distribution
print("\nFeedback Distribution:")
print(df["Feedback"].value_counts())

# Top Pin Codes
print("\nTop Pin Codes by Customer Count:")
print(df["Pin code"].value_counts())

# Correlation between numerical columns
print("\nCorrelation Matrix:")
print(df.corr(numeric_only=True))

##HERE IS SOME BUSSINESS INSIGHTS BY USING ONLINE FOOD DATASET


#1]The majority of customers belong to the most common occupation.
#2]Most customers fall into a specific monthly income category.
#3]Positive feedback is higher than negative feedback, indicating good customer satisfaction.
#4]A few Pin Codes contribute the highest number of customers, showing high-demand areas.
#5]Numerical features with strong correlations can help the company understand customer behavior.
```
